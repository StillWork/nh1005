# 사용자-아이템 점수 테이블
- 사용자-영화-점수 테이블을 이용
- pivot_table 이용

```python
ratings = pd.read_csv('https://raw.githubusercontent.com/StillWork/data/master/ratings.csv')
movies = pd.read_csv('https://raw.githubusercontent.com/StillWork/data/master/movies.csv')
```

```python
print(ratings.shape)
ratings[:3]
```

```python
ratings.userId.value_counts()
```

```python
print(movies.shape)
movies[:3]
```

```python
ratings_movies = pd.merge(ratings, movies, on = 'movieId')
print(ratings_movies.shape)
ratings_movies[:3]
```

- pivot table을 이용한다

```python
user_movie_rating = ratings_movies.pivot_table('rating', index = 'userId', columns='title')
user_movie_rating
```

## 아이템-사용자 테이블


```python
# movie_user_rating = ratings_movies.pivot_table('rating', index = 'title', columns='userId')
movie_user_rating = user_movie_rating.T
movie_user_rating
```

```python
movie_user_rating.fillna(0, inplace = True)
movie_user_rating.head(3)
```

# 아이템 기반 협업 필터링

- 아이템 기반 협업 필터링은 비슷한 영화를 찾아서 추천하는 것
- 비슷한 영화를 코사인 유사도로 측정한다 (행 단위로(아이템 단위로) 상호 유사도를 측정한다)
 - 사용자 Id를 컬럼으로 하는 특성 공간상의 벡터로 표현한다

```python
item_based_collabor = cosine_similarity(movie_user_rating)
print(item_based_collabor.shape)
item_based_collabor[:3]
```

```python
item_based_collabor = pd.DataFrame(data = item_based_collabor, 
   index = movie_user_rating.index, columns = movie_user_rating.index)
item_based_collabor[:3]
```

```python
print(movie_user_rating.shape)
print(item_based_collabor.shape)
```

- 특정 item에 대해 유사도가 높은 순으로 10개를 찾는 함수
- 인덱스가 title임

```python
def get_item_based_collabor(title):
    return item_based_collabor[title].sort_values(ascending=False)[1:11]
```

```python
get_item_based_collabor('Godfather, The (1972)')
```

```python
get_item_based_collabor('American Beauty (1999)')
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

## 유클리디언 거리 사용
- 코사인 유사도가 아닌 특성 공간상의 직선거리를 사용하는 경우

```python
item_based_distance= euclidean_distances(movie_user_rating,
                                         movie_user_rating)
```

```python
item_based_distance = pd.DataFrame(data = item_based_distance, 
   index = movie_user_rating.index, columns = movie_user_rating.index)
item_based_distance[:3]
```

```python
def get_item_based_distance(title):
    return item_based_distance[title].sort_values(ascending=True)[1:11]
```

```python
get_item_based_distance('Godfather, The (1972)')
```

```python
get_item_based_distance('American Beauty (1999)')
```
